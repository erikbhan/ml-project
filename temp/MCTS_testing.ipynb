```python
import matplotlib.pyplot as plt
import time

from temp.src.Node import Node
```

```python
n, avg_times = [2**i for i in range(1, 19)], []

print(f"Running {len(n)} simulations:")
for i in range(len(n)):
    root = Node(parent=None, env=None, move=None)
    start = time.time()
    root.mcts(n[i])
    stop = time.time()
    avg_times.append(((stop-start)/n[i]) * 1000)
    print(f"{i+1:3d}: {n[i]:7d} iterations took {stop - start:10.4f} sec...")
```

```python
fig, ax = plt.subplots()
ax.loglog(n, avg_times)
ax.set_xlabel("iterations")
ax.set_ylabel("milliseconds per iteration")
plt.show()
```
